### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
gmaps.configure(api_key=g_key)
```

```python
# Create a marker_layer using the humidity list to fill the info box

marker_locations = weather_df[['Lat', 'Lng']]

humidity = weather_df["Humidity"].astype(float)

fig = gmaps.figure()
heatmap_df = gmaps.heatmap_layer(marker_locations, weights=humidity,
                                   dissipating=False, max_intensity=5,
                                   point_radius=1)

fig.add_layer(heatmap_df)

fig
```
